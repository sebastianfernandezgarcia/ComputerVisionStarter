```python
#TAREA: exportamos cada canal en una imagen distinta.
import cv2
import numpy as np

cap = cv2.VideoCapture(0)
cv2.namedWindow('Webcam',cv2.WINDOW_NORMAL)
zeros = None
while True:
    ret, frame = cap.read()

    if ret == True:
        # split the image into its RGB channels
        h, w, c = frame.shape #height, width, capas
        img_vacia = np.zeros((h, w), dtype="uint8") #crear imagen vacia del tamaño que devuelve la cámara (el frame)


        #En OpenCV, por motivo raro, el orden es B,G,R
        (B, G, R) = cv2.split(frame)

        #en cada componente ponemos solo su canal
        B = cv2.merge([B, img_vacia, img_vacia])
        G = cv2.merge([img_vacia, G, img_vacia])
        R = cv2.merge([img_vacia, img_vacia, R])

        #Vamos a poner 4 imagenes entonces necesitamos doble de ancho y alto
        final = np.zeros((h * 2, w * 2, 3), dtype="uint8")

        final[0:h, 0:w] = frame # de 0 hasta 1h ponemos el original
        final[0:h, w:w * 2] = B # de 0 a h en el segundo de w ponemos el azul
        final[h:h * 2, 0:w] = G   # abajo izquirda el verde
        final[h:h * 2, w:w * 2] = R  # abajo derecha el rojo

        cv2.imshow('Webcam', final)
        #cv2.imshow('Cam3', cv2.resize(collage, (int(w*1.5),int(h/2)),cv2.INTER_NEAREST))
    
    else:
        break

    if cv2.waitKey(20) == 27:
            break
cap.release()
cv2.destroyAllWindows()



```

```python
#TAREA: exportamos cada canal en una imagen distinta pero jugamos con los colores y modificamos.
import cv2
import numpy as np

cap = cv2.VideoCapture(0)
cv2.namedWindow('Webcam',cv2.WINDOW_NORMAL)
zeros = None
while True:
    ret, frame = cap.read()

    if ret == True:
        # split the image into its RGB channels
        h, w, c = frame.shape #height, width, capas
        img_vacia = np.zeros((h, w), dtype="uint8") #crear imagen vacia del tamaño que devuelve la cámara (el frame)


        #En OpenCV, por motivo raro, el orden es B,G,R
        (b, g, r) = cv2.split(frame)

        #en cada componente ponemos solo su canal
        B = cv2.merge([g-b, b, r])  #arbitrariamente sumamos o restamos valores en otros ordenes
        G = cv2.merge([r, b, g-r])
        R = cv2.merge([b, r, g-r])

        #Vamos a poner 4 imagenes entonces necesitamos doble de ancho y alto
        final = np.zeros((h * 2, w * 2, 3), dtype="uint8")

        final[0:h, 0:w] = frame # de 0 hasta 1h ponemos el original
        final[0:h, w:w * 2] = B # de 0 a h en el segundo de w ponemos el azul
        final[h:h * 2, 0:w] = G   # abajo izquirda el verde
        final[h:h * 2, w:w * 2] = R  # abajo derecha el rojo

        cv2.imshow('Webcam', final)
    
    else:
        break

    if cv2.waitKey(20) == 27:
            break
cap.release()
cv2.destroyAllWindows()
```

Evento de ratón para mostrar el valor RGB del píxel

```python
# Función que trata eventos de ratón 
def mouse_events(event, x, y, flags, params):
    global px, py, R, G, B
    px = -1
    # Movimiento del puntero     
    if event==cv2.EVENT_MOUSEMOVE:
        #Valores del píxel
        B = frame[y, x, 0]
        G = frame[y, x, 1]
        R = frame[y, x, 2]  

        px = x
        py = y

     
  
```

```python
vid = cv2.VideoCapture(0)

# Tipografía
font = cv2.FONT_HERSHEY_SIMPLEX
  
lanzado = 0
px = -1
while(True):      
    # fotograma a fotograma
    ret, frame = vid.read()

    if ret: 
        # Activa em manejador en el primer fotograma
        if lanzado == 0:
            # Muestra fotograma
            cv2.imshow('Cam', frame)   
            # Define el nombre del manejador del evento
            cv2.setMouseCallback('Cam', mouse_events)
            lanzado = 1
            
        # Muestra valores RGB intentando centrar en el puntero
        if px > -1:
            cv2.putText(frame, '{}'.format(R), (px-45,py-5), font, 0.5, (0, 0, 255), 1)
            cv2.putText(frame, '     {}'.format(G), (px-54,py-5), font, 0.5, (0, 255, 0), 1)
            cv2.putText(frame, '         {}'.format(B), (px-54,py-5), font, 0.5, (255, 0, 0), 1)
            
        cv2.imshow('Cam', frame)   
        
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

```python
#ESTO ES CODIGO QUE ESTUVE VIENDO POR AHI PARA HACER Y PROBAR MAS COSAS
# Fuente: https://docs.opencv.org/4.x/db/d8e/tutorial_threshold.html
import cv2
import numpy as np
# Setup camera
cap = cv2.VideoCapture(0)
# Set a smaller resolution
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

def nothing(x):
    pass

canny = "Canny"
cv2.namedWindow(canny)
cv2.createTrackbar('Threshold 1', canny, 0, 255, nothing)
cv2.createTrackbar('Threshold 2', canny, 0, 255, nothing)
while True:
    _, frame = cap.read()
    frame = cv2.flip(frame, 1)
    t1 = cv2.getTrackbarPos('Threshold 1', canny)
    t2 = cv2.getTrackbarPos('Threshold 2', canny)
    gb = cv2.GaussianBlur(frame, (5, 5), 0)
    can = cv2.Canny(gb, t1, t2)
    cv2.imshow(canny, can)
    frame[np.where(can)] = 255
    cv2.imshow('WebCam', frame)
    if cv2.waitKey(1) == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
vid = cv2.VideoCapture(0)

#Dimensiones de la cámara
w = int(vid.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))

#Fuerzo a mitad de resolución para ocupar menos pantalla
w=int(w/2)
h=int(h/2)
vid.set(cv2.CAP_PROP_FRAME_WIDTH, w)
vid.set(cv2.CAP_PROP_FRAME_HEIGHT, h)

#Imagen conjunta 2x original
collage = np.zeros((h*2,w*2,3), dtype = np.uint8)
tl = collage[0:h,0:w]
tr = collage[0:h,w:w+w]
bl = collage[h:h+h,0:w]
br = collage[h:h+h,w:w+w]

while True:      
    # fotograma a fotograma
    ret, frame = vid.read()

    if ret:
        #Separamos canales
        r = frame[:,:,0]
        g = frame[:,:,1]
        b = frame[:,:,2]

        #Jugamos
        tl[:,:,0] = r
        tl[:,:,1] = g
        tl[:,:,2] = b

        tr[:,:,0] = 255 - g
        tr[:,:,1] = r
        tr[:,:,2] = b
        
        bl[:,:,0] = b
        bl[:,:,1] = 255 - r
        bl[:,:,2] = g

        br[:,:,0] = r
        br[:,:,1] = g
        br[:,:,2] = 255 - b
    
        # Muestra composicion
        cv2.imshow('Cam', collage)
    
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

Transformando la imagen capturada en una colección de círculos

```python
vid = cv2.VideoCapture(0)

ncells = 10
off = int(ncells/2)
  
while(True):      
    # fotograma a fotograma
    ret, frame = vid.read()

    if ret:
        #Dimensiones originales
        h, w, c = frame.shape
        #Redimensiono
        down_frame = cv2.resize(frame, (int(w/ncells),int(h/ncells)),cv2.INTER_NEAREST)
        #Dimensiones reducidas
        h2, w2, c2 = down_frame.shape

        #Separamos canales
        r = down_frame[:,:,0]
        g = down_frame[:,:,1]
        b = down_frame[:,:,2]

        # Creamos imagen negra
        gris_up_frame = np.zeros((h2*ncells,w2*ncells,1), dtype = np.uint8)

        for y in range(0,h2):
            for x in range(0,w2):
                # La suma de los valores RGB define el radio del círculo
                rad = int ( (r[y,x] + g[y,x] + b[y,x])/(ncells*3*2))
                cv2.circle(gris_up_frame,(x*ncells+off, y*ncells + off), rad, (255,255,255), -1)

    
        # Muestra fotograma redimensionando a la mitad para que quepa en pantalla
        cv2.imshow('Cam', gris_up_frame)
    
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```
